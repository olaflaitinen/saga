# Notebook 01: Conformal Calibration

This notebook demonstrates the split conformal calibration procedure (Theorem 1) and
the Adaptive Temporal Conformal Prediction bound (Theorem 2) on the synthetic mirror.

```python
import numpy as np
from saga.conformal.adaptive_temporal import AdaptiveTemporalConformalCalibrator
from saga.evaluation.metrics_probabilistic import picp

rng = np.random.default_rng(20260601)
n_cal = 14_107
y_cal = rng.normal(0, 1, n_cal)
q_lower_cal = y_cal - 1.28
q_upper_cal = y_cal + 1.28

calibrator = AdaptiveTemporalConformalCalibrator(alpha=0.10, lipschitz_constants={10: 0.65})
q_hat = calibrator.calibrate(q_lower_cal, q_upper_cal, y_cal, horizon=10)
bound = calibrator.theorem2_bound(horizon=10, delta=0.10)

print(f'Q-hat at h=10: {q_hat:.4f}')
print(f'Theorem 2 worst-case deviation bound: {bound:.4f}')
print(f'Manuscript bound (Table III): ~0.024')
```
