# Notebook 04: Ablation Study

Summarizes the 13 ablation variants (A1-A13, Table VI) and their manuscript CRPS values.

```python
import pandas as pd
from saga.evaluation.ablation_runner import _ABLATION_SPECS

rows = [
    {'Label': s.label, 'Description': s.description, 'CRPS (h=10)': s.crps_manuscript}
    for s in _ABLATION_SPECS
]
df = pd.DataFrame(rows)
df['Delta CRPS'] = df['CRPS (h=10)'] - df.loc[df['Label'] == 'headline', 'CRPS (h=10)'].item()
print(df.to_string(index=False))
```
