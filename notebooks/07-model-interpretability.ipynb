# Notebook 07: Model Interpretability

Demonstrates attention weight aggregation and integrated gradients attribution
on a small synthetic batch. Reproduces the qualitative findings of Section V-G.

```python
import torch
from saga.config import SagaConfig
from saga.model.saga_model import SagaModel
from saga.interpretability.integrated_gradients import IntegratedGradients

config = SagaConfig()
model = SagaModel(config)
model.eval()

ig = IntegratedGradients(model, n_steps=10)

continuous = torch.randn(1, 10, 15)
categorical = {n: torch.zeros(1, 10, dtype=torch.long) for n in [
    'occupation', 'industry', 'region', 'education_level',
    'field_of_study', 'sex', 'country_of_birth_group',
    'marital_status', 'n_children', 'age_youngest_child'
]}
missingness = torch.zeros(1, 10, 16)
age = torch.full((1, 10), 30, dtype=torch.long)
year = torch.full((1, 10), 2005, dtype=torch.long)

attrs = ig.attribute(continuous, categorical, missingness, age, year)
print(f'Integrated gradients shape: {attrs.shape}')
print(f'Most important feature at last position: index {attrs[-1].abs().argmax()}')
```
