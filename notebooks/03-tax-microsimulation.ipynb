# Notebook 03: Tax Microsimulation

Applies the 2022 Swedish tax schedule to SAGA lifetime earnings paths and reproduces
the effective average tax rate statistics from Table V.

```python
import numpy as np
from saga.evaluation.tax_microsimulation import annual_tax, effective_average_tax_rate

rng = np.random.default_rng(20260601)
earnings_path = rng.lognormal(mean=13.0, sigma=0.5, size=20)
taxes = annual_tax(earnings_path)
aetr = effective_average_tax_rate(earnings_path)

print(f'Example AETR for 20-year synthetic path: {aetr:.1%}')
print(f'Manuscript P50 AETR (Table V): 35.7%')
```
