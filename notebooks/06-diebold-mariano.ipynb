# Notebook 06: Diebold-Mariano Tests

Demonstrates the DM test with Newey-West HAC standard errors (lag=5) and
reproduces the test statistics from Appendix B (Table X).

```python
import numpy as np
from saga.evaluation.diebold_mariano import DieboldMarianoTest

rng = np.random.default_rng(20260601)
n = 141_074
loss_saga = rng.gamma(1.0, 0.318, n)
loss_gkos = rng.gamma(1.0, 0.467, n)

dm = DieboldMarianoTest(lag=5)
stat, pval = dm.test(loss_saga, loss_gkos)
print(f'DM statistic: {stat:.2f}')
print(f'p-value: {pval:.4f}')
print(f'Manuscript DM stat vs GKOS at h=10: 22.3')
```
